10. What is the difference between implicit and explicit type casting?
    - In programming, "implicit type casting" means a data type conversion done automatically by the compiler without any explicit instruction from the programmer, usually when converting a smaller data type to a larger one, while "explicit type casting" requires the programmer to manually specify the conversion using a casting operator, typically when there might be potential data loss by converting to a smaller data type; essentially, implicit casting happens automatically, while explicit casting is done intentionally by the developer.

        print("Could not convert to float.")

    if input_string.lower() in ["true", "1"]:
        boolean_value = True
    elif input_string.lower() in ["false", "0"]:
        boolean_value = False
    else:
        boolean_value = None  # Handle cases that are not clearly True or False

    if boolean_value is not None:
        print("Boolean:", boolean_value)
    else:
        print("Could not convert to boolean.")


if __name__ == "__main__":
    convert_to_types()

14. Write a program to reverse a string using a while loop.
    - def reverse_string(text):
    """Reverses a string using a while loop.

    Args:
        text: The string to reverse.

    Returns:
        The reversed string.
    """
    reversed_text = ""
    index = len(text) - 1  # Start from the last index

    while index >= 0:
        reversed_text += text[index]  # Add the current character to the reversed string
        index -= 1  # Move to the previous index

    return reversed_text


# Get input from the user
string_to_reverse = input("Enter a string: ")

# Reverse the string and print the result
reversed_string = reverse_string(string_to_reverse)
print("Reversed string:", reversed_string)

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
    - def factorial(number):
    """Calculates the factorial of a number using a while loop.

    Args:
        number: The number for which to calculate the factorial.

    Returns:
        The factorial of the number.
    """

    if number < 0:
        return "Factorial is not defined for negative numbers."
    elif number == 0:
        return 1
    else:
        factorial = 1
        while number > 1:
            factorial *= number
            number -= 1
        return factorial


# Get input from the user
num = int(input("Enter a non-negative integer: "))

# Calculate and print the factorial
result = factorial(num)
print(f"The factorial of {num} is {result}.")
